```python
## Import necessary library.
import pandas as pd

## Read the .csv file.
df = pd.read_csv(r'C:\Users\hp\kaggle\.kaggle\athlete_events.csv')
```

# Filtering data from the existing dataframe.

```python
## Create a dataframe with only male athletes in 2016 Summer Olympics Swimming Men's 100 metres Freestyle Competiton.
male_df = df[(df['Year'] == 2016) & 
            (df['Sport'] == 'Swimming') & 
            (df['Season'] == 'Summer') & 
            (df['Event'] == "Swimming Men's 100 metres Freestyle")]

male_df.head()
```

```python
## Create a dataframe with only female athletes in 2016 Summer Olympics Swimming Women's 100 metres Freestyle Competiton.
female_df = df[(df['Year'] == 2016) & 
            (df['Sport'] == 'Swimming') & 
            (df['Season'] == 'Summer') & 
            (df['Event'] == "Swimming Women's 100 metres Freestyle")]

female_df.head()
```

```python
## Concatenate the two dataframes to create a dataframe of all athletes.
all_athletes_df = pd.concat([male_df, female_df], ignore_index = True)
all_athletes_df.head()
```

# Joining athlete speed with height.

```python
## Read the .csv file of athletes and their completion times obtained from Wikipedia.
all_athletes_time_df = pd.read_csv(r'C:\Users\hp\kaggle\.kaggle\all_athletes_time.csv', encoding='latin1')

## Calculate speed of all athletes.
all_athletes_time_df['Speed'] = 100 / all_athletes_time_df['Time'] ## Divide the total distance covered (100m) by time taken.
all_athletes_time_df
```

```python
## Create a dataframe of only male athletes.
male_athletes_time = all_athletes_time_df[(all_athletes_time_df['Gender'] == 'Male')]
male_athletes_time.head()
```

```python
## Create a dataframe of only female athletes.
female_athletes_time = all_athletes_time_df[(all_athletes_time_df['Gender'] == 'Female')]
female_athletes_time.head()
```

```python
## Join the dataframe of male athletes with the dataframe of their heights.
male_athletes_v1 = pd.merge (male_athletes_time, male_df, on = 'Name', how = 'left') # Join the dataframes.
male_selected_columns = ['Height', 'Speed'] # Create a variable with selected columns. 
male_athletes_v2 = male_athletes_v1[male_selected_columns] #Create a filtered dataframe.
male_athletes_complete = male_athletes_v2.dropna(subset=['Height']) # Create a final dataframe with all null values dropped.
male_athletes_complete.head()
```

```python
## Join the dataframe of female athletes with the dataframe of their heights.
female_athletes_v1 = pd.merge (female_athletes_time, female_df, on = 'Name', how = 'left') # Join the dataframes.
female_selected_columns = ['Height', 'Speed'] # Create a variable with selected columns. 
female_athletes_v2 = female_athletes_v1[female_selected_columns] #Create a filtered dataframe.
female_athletes_complete = female_athletes_v2.dropna(subset=['Height']) # Create a final dataframe with all null values dropped.
female_athletes_complete.head()
```

```python
## Join the dataframe of all athletes with the dataframe of their heights.
all_athletes_v1 = pd.merge (all_athletes_time_df, all_athletes_df, on = 'Name', how = 'left') # Join the dataframes.
all_athletes_selected_columns = ['Height', 'Speed'] # Create a variable with selected columns. 
all_athletes_v2 = all_athletes_v1[all_athletes_selected_columns] #Create a filtered dataframe.
all_athletes_complete = all_athletes_v2.dropna(subset=['Height']) # Create a final dataframe with all null values dropped.
all_athletes_complete.head()
```

# Visualizing results.

```python
## Importing necessary libraries.
import matplotlib.pyplot as plt
import numpy as np
```

### Visualizing results of all athletes.

```python
## Create a scatter plot.
plt.figure(figsize=(8,5))
plt.scatter(all_athletes_complete['Height'], all_athletes_complete['Speed'], color = 'red', label = 'Height vs Speed')

## Calculate a line of best fit.
slope, intercept = np.polyfit(all_athletes_complete['Height'], all_athletes_complete['Speed'], 1)

## Generate a line of best fit.
all_athletes_bfl = slope * all_athletes_complete['Height'] + intercept

## Plot the line of best fit.
plt.plot (all_athletes_complete['Height'], all_athletes_bfl, color = 'r')

## Add labels and title.
plt.xlabel('Height(cm)')
plt.ylabel('Speed(m/s)')
plt.title('Relationship between Height and Speed of all athletes in Heats of Swimming 100m Freestyle, 2016 Olympics')
plt.show()
```

### Visualizing results of male athletes.

```python
## Create a scatter plot.
plt.figure(figsize=(8,5))
plt.scatter(male_athletes_complete['Height'], male_athletes_complete['Speed'], color = 'blue', label = 'Height vs Speed')

## Calculate a line of best fit.
slope, intercept = np.polyfit(male_athletes_complete['Height'], male_athletes_complete['Speed'], 1)

## Generate a line of best fit.
male_athletes_bfl = slope * male_athletes_complete['Height'] + intercept

## Plot the line of best fit.
plt.plot (male_athletes_complete['Height'], male_athletes_bfl, color = 'blue')

## Add labels and title.
plt.xlabel('Height(cm)')
plt.ylabel('Speed(m/s)')
plt.title('Relationship between Height and Speed of male athletes in Heats of Swimming 100m Freestyle, 2016 Olympics')
plt.show()
```

### Visualizing results of female athletes.

```python
## Create a scatter plot.
plt.figure(figsize=(8,5))
plt.scatter(female_athletes_complete['Height'], female_athletes_complete['Speed'], color = 'green', label = 'Height vs Speed')

## Calculate a line of best fit.
slope, intercept = np.polyfit(female_athletes_complete['Height'], female_athletes_complete['Speed'], 1)

## Generate a line of best fit.
female_athletes_bfl = slope * female_athletes_complete['Height'] + intercept

## Plot the line of best fit.
plt.plot (female_athletes_complete['Height'], female_athletes_bfl, color = 'green')

## Add labels and title.
plt.xlabel('Height(cm)')
plt.ylabel('Speed(m/s)')
plt.title('Relationship between Height and Speed of female athletes in Heats of Swimming 100m Freestyle, 2016 Olympics')
plt.show()
```
